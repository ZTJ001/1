```python
#model.py
from django.db import models

class BlogPost(models.Model):
    """创建BlogPost模型"""
    title = models.CharField(max_length=200)
    text = models.TextField()
    date_added = models.DateTimeField(auto_now_add=True)

    def __str__(self):
        """返回模型的字符串显示"""
        if len(self.title) < 50:
            return self.title
        else:
            return self.title[:50]+"..."
        if len(self.text) < 200:
            return self.text
        else:
            return self.text[:200]+"..."

```

```python
#views.py-2 添加与编辑
from django.shortcuts import render
from django.http import HttpResponseRedirect
from django.core.urlresolvers import reverse


from .models import BlogPost
from .forms import BlogForm,TextForm

def index(request):
    """主页内容"""
    return render(request,'blogs/index.html')

def blogs(request):
    """显示所有blog名"""
    blogs = BlogPost.objects.order_by('date_added')
    context = {'blogs':blogs}
    return render(request,'blogs/blogs.html',context)

def blog(request,blog_id):
    """显示每个blog的内容"""
    blog = BlogPost.objects.get(id=blog_id)
    text = blog.text
    context= {'blog':blog,'text':text}
    return render(request,'blogs/blog.html',context)

def new_blog(request):
    """添加新Blog"""
    if request.method != 'POST':
        #创建新表单
        form = BlogForm()
    else:
        #Post提交数据进行处理
        form = BlogForm(request.POST)
        if form.is_valid():
            form.save()
            return HttpResponseRedirect(reverse('blogs:blogs'))
    context = {'form':form}
    return render(request,'blogs/new_blog.html',context)

def edit_blog(request,blog_id):
    """编辑现有Blog"""
    text = BlogPost.text.objects.get(id=blog_id)
    blog = text.blog
    if request.method != 'POST':
        #创建新表单
        form = TextForm(instance=text)
    else:
        #Post提交数据进行处理
        form = TextForm(instance=text,data=request.POST)
        if form.is_valid():
            form.save()
            return HttpResponseRedirect(reverse('blogs:blog',args=[blog.id]))
    context = {'text':text,'blog':blog,'form':form}
    return render(request,'blogs/edit_blog.html',context)
```
